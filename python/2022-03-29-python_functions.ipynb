## Python Useful Built-in Functions

### Input/Output

```python
# Get user input
input("Enter some text:")
```

```python
# Open file and return file object
open("python.ipynb", "r")
```

```python
# Print to standard output
print("Hello, World!")
```

```python
# Return readable version of object
repr("a\n")
```

```python
# Evaluate and execute expression
eval('1 + 1')
```

```python
# Return list of object's properties and methods
dir(str)
```

### Casting

```python
# Return boolean value of object
bool([])
```

```python
# Return integer number value of object
int(1.1)
```

```python
# Return floating point number value of object
float(1)
```

```python
# Return string value of object
str(-1)
```

```python
# Return list value of object
list(range(3))
```

```python
# Return tuple value of object
tuple([1, 2, 3])
```

```python
# Return dictionary value of object
dict(a=1, b=2, c=3)
```

### Arithmetic

```python
# Return absolute value of number
abs(-1)
```

```python
# Return value of x to the power of y
pow(2, 3)
```

```python
# Return rounded value of number
round(1.1)
```

```python
# Return quotient and remainder when arg1 is divided by arg2
divmod(5, 2)
```

### Equality

```python
# Return True if arg1 is instance of arg2
isinstance('a', int)
```

```python
# Return True if arg1 is subclass of arg2
issubclass(list, dict)
```

### Iterables

#### Create

```python
# Return length of object
len([1, 2, 3])
```

```python
# Return sequence of numbers, by default starting from 0 and increments of 1
list(range(5, 10, 2))
```

```python
# Return reversed iterator of sequence
list(reversed([1, 2, 3]))
```

```python
# Return sorted list of sequence
sorted(["a", "aa", "aaa"], key=lambda item: len(item), reverse=True)
```

```python
# Return enumerate object from iterable, adds counter alongside items
type(enumerate(["a", "b", "c"]))
```

```python
# Return iterator from two or more iterables, aggregates together (up to longest iterable)
list(zip(["a", "b", "c"], [1, 10, 100, 1000, 10000]))
```

#### Modify

```python
# Return sum of items in iterable
sum([1, 2, 3])
```

```python
# Return smallest item in iterable
min([1, 2, 3])
```

```python
# Return largest item in iterable
max([1, 2, 3])
```

```python
# Return True if all items in iterable are true
all([True, False])
```

```python
# Return True if any item in iterable is true
any([True, False])
```

```python
# Returns iterator, applies map function to each item in iterable
list(map(lambda item: 10 ** item, [1, 2, 3]))
```

```python
# Returns iterator, applies filter function to exclude items in iterable
list(filter(lambda item: item != 1, [1, 2, 3]))
```

```python
# Returns iterator, applies reduce function to each item in iterable
from functools import reduce
reduce(lambda acc, item: acc * item, [1, 2, 3], 1)
```
